### Thresholded binary image creation

```python
# GrayScale Binary Image Function

def gray_select(img, thresh = (180, 255)):
    
    # Convert to GRAYSCALE color space
    #gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # create binary image
    binary_output = np.zeros_like(gray)
    binary_output[(gray > thresh[0]) & (gray <= thresh[1])] = 1
    
    # Return a binary image of threshold result
    return binary_output
    
```

```python
# RGB Binary Image Function - Red

def rgb_select(img, thresh = (200, 255)):
    
    # apply threshold to the R channel (BGR image read in using opencv)
    R = img[:,:,2]
    
    # create binary image
    binary_output = np.zeros_like(R)
    binary_output[(R > thresh[0]) & (R <= thresh[1])] = 1
    
    # Return a binary image of threshold result
    return binary_output
    
```

```python
# HLS Binary Function - Saturation

def hls_select(img, thresh=(90, 255)):
    # Convert to HLS color space
    hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
    # Apply a threshold to the Saturation channel
    S = hls[:,:,2]
    
    # create binary image
    binary_output = np.zeros_like(S)
    binary_output[(S > thresh[0]) & (S <= thresh[1])] = 1
    
    # Return a binary image of threshold result
    return binary_output
```

```python
# HSV Binary Function - Value

def hsv_select(img, thresh=(90, 100)):
    # Convert to HLS color space
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    # Apply a threshold to the Saturation channel
    V = hsv[:,:,2]
    
    # create binary image
    binary_output = np.zeros_like(V)
    binary_output[(V> thresh[0]) & (V <= thresh[1])] = 1
    
    # Return a binary image of threshold result
    return binary_output
```

```python
# LUV Binary Function - V Channle 

def luv_select(img, thresh=(225, 255)):
    # Convert to LUV color space
    luv = cv2.cvtColor(img, cv2.COLOR_BGR2LUV)
    # Apply a threshold to the L channel
    V = luv[:,:,2]

    # create binary image
    binary_output = np.zeros_like(V)
    binary_output[(V >= thresh[0]) & (V <= thresh[1])] = 1

    # Return a binary image of threshold result
    return binary_output
```

```python
# Lab Binary Function - b - to get yellow line

def lab_select(img, thresh=(75, 100)):
    # Convert to Lab color space
    lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
    # Apply a threshold to the b channel
    l = lab[:,:,0]
    a = lab[:,:,1]
    b = lab[:,:,2]

    # create binary image
    binary_output = np.zeros_like(b)
    binary_output[(l > 80) & (b > 60) & (a < 10 ) & (a >= -36)] = 1

    # Return a binary image of threshold result
    return binary_output
```

```python
# YUV Binary Function - Red
def yuv_select(img, thresh=(0.9, 1.0)):
    # Convert to YUV color space
    yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
    # apply threshold to the R channel
    y = yuv[:,:,0] # luma - brightness
    u = yuv[:,:,1] # Chrominance Pr
    v = yuv[:,:,2] # Chrominance Pb
    
    # create binary image
    binary_output = np.zeros_like(y)
    binary_output[(y >= thresh[0]) & (y <= thresh[1])] = 1
    
    # Return a binary image of threshold result
    return binary_output
    
    
```

```python
# Sobel X gradient

def abs_sobel_thresh(img, orient='x', thresh=(20, 100)):

    #convert image to grayscale ready for sobel
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply cv2.Sobel()
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    
    # Take the absolute value of the output from cv2.Sobel()
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    if orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    
    # Scale the result to an 8-bit range (0-255)
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    # Apply lower and upper thresholds
    sx_binary_output = np.zeros_like(scaled_sobel)
    sx_binary_output[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    
    # Create binary_output
    return sx_binary_output
```

```python
# Sobel gradient to edges
def dir_threshold(img, sobel_kernel=3, thresh=(0, np.pi/2)):
    # Calculate gradient direction
    
    #convert image to grayscale ready for sobel
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply cv2.Sobel() - get x gradient
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    # Apply cv2.Sobel() - get x gradient
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    
    #Use np.arctan2(abs_sobely, abs_sobelx) to calculate the direction of the gradient 
    grad_dir = np.arctan2(np.absolute(sobely), np.absolute(sobelx))
    
    # Apply threshold
    # Create a binary mask where direction thresholds are met
    dir_binary = np.zeros_like(grad_dir)
    dir_binary[(grad_dir >= thresh[0]) & (grad_dir <= thresh[1])] = 1
    
    return dir_binary
```

```python
# Sobel magnitude threshold  

def mag_thresh(img, sobel_kernel=3, mag_thresh=(0, 255)):
    # Calculate gradient magnitude
    
    #convert image to grayscale ready for sobel
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Apply cv2.Sobel() - get x gradient
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
    # Apply cv2.Sobel() - get x gradient
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
    
    # Take the absolute value of the output from cv2.Sobel()
    # get magnitude of gradient x
    abs_sobelx = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    # get magnitude of gradient x
    abs_sobely = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    
    abs_sobelxy = np.sqrt(abs_sobelx**2+abs_sobely**2)
    
    # Scale the result to an 8-bit range (0-255)
    # Rescale to 8 bit
    scale_factor = np.max(abs_sobelxy)/255 
    abs_sobelxy = (abs_sobelxy/scale_factor).astype(np.uint8) 
    
    # Apply threshold
    mag_binary = np.zeros_like(abs_sobelxy)
    mag_binary[(abs_sobelxy >= mag_thresh[0]) & (abs_sobelxy <= mag_thresh[1])] = 1
    
    return mag_binary
  
```
